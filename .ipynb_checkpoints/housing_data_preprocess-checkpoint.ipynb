```python
# import dependencies
import numpy as np
import pandas as pd
import datetime
import warnings
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load Ames_HousePrice.csv data
housing_data = pd.read_csv('Ames_HousePrice.csv', index_col=0)
```

```python
#1) inspect features
header_lst = housing_data.columns
for header in header_lst:
    print(header)
    print('missing values: ' + str(housing_data[header].isnull().sum()))
    print('data type: ' + str(housing_data[header].dtype))

# initial sum of all missing values
missing_sum_init = sum(housing_data.isnull().sum())
# print('initial sum of all missing values: ' + str(missing_sum_init))
```

```python
#2) replace missing values (refer to data dictionary for details)
housing_data['LotFrontage'] = housing_data['LotFrontage'].fillna(housing_data['LotFrontage'].mean())
housing_data['Alley'] = housing_data['Alley'].fillna('NA')
housing_data['MasVnrType'] = housing_data['MasVnrType'].fillna('None')
housing_data['MasVnrArea'] = housing_data['MasVnrArea'].fillna(0)
housing_data['BsmtQual'] = housing_data['BsmtQual'].fillna('NA')
housing_data['BsmtCond'] = housing_data['BsmtCond'].fillna('NA')
housing_data['BsmtExposure'] = housing_data['BsmtExposure'].fillna('NA')
housing_data['BsmtFinType1'] = housing_data['BsmtFinType1'].fillna('NA')
housing_data['BsmtFinSF1'] = housing_data['BsmtFinSF1'].fillna(0)
housing_data['BsmtFinType2'] = housing_data['BsmtFinType2'].fillna('NA')
housing_data['BsmtFinSF2'] = housing_data['BsmtFinSF2'].fillna(0)
housing_data['BsmtUnfSF'] = housing_data['BsmtUnfSF'].fillna(0)
housing_data['TotalBsmtSF'] = housing_data['TotalBsmtSF'].fillna(0)
housing_data['Electrical'] = housing_data['Electrical'].fillna('Mix')
housing_data['BsmtFullBath'] = housing_data['BsmtFullBath'].fillna(0)
housing_data['BsmtHalfBath'] = housing_data['BsmtHalfBath'].fillna(0)
housing_data['FireplaceQu'] = housing_data['FireplaceQu'].fillna('NA')
housing_data['GarageType'] = housing_data['GarageType'].fillna('NA')
housing_data['GarageYrBlt'] = housing_data['GarageYrBlt'].fillna(0)
housing_data['GarageFinish'] = housing_data['GarageFinish'].fillna('NA')
housing_data['GarageCars'] = housing_data['GarageCars'].fillna(0)
housing_data['GarageArea'] = housing_data['GarageArea'].fillna(0)
housing_data['GarageQual'] = housing_data['GarageQual'].fillna('NA')
housing_data['GarageCond'] = housing_data['GarageCond'].fillna('NA')
housing_data['PoolQC'] = housing_data['PoolQC'].fillna('NA')
housing_data['GarageCond'] = housing_data['GarageCond'].fillna('NA')
housing_data['Fence'] = housing_data['Fence'].fillna('NA')
housing_data['MiscFeature'] = housing_data['MiscFeature'].fillna('NA')

# post sum of all missing values
missing_sum_post = sum(housing_data.isnull().sum())
# print('post process sum of all missing values: ' + str(missing_sum_post))
```

```python
# 3) enocde categorical ordinal
# define scale mapper dictionary for categorical ordinal features
scale_mapper_ExterQual = ['ExterQual', {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'NA':0}]
scale_mapper_ExterCond = ['ExterCond', {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'NA':0}]
scale_mapper_BsmtQual = ['BsmtQual', {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'NA':0}]
scale_mapper_ExterCond = ['ExterCond', {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'NA':0}]
scale_mapper_BsmtQual = ['BsmtQual', {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'NA':0}]
scale_mapper_BsmtCond = ['BsmtCond', {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'NA':0}]
scale_mapper_BsmtExposure = ['BsmtExposure', {'Gd':4, 'Av':3, 'Mn':2, 'No':1, 'NA':0}]
scale_mapper_BsmtFinType1 = ['BsmtFinType1', {'GLQ':6, 'ALQ':5, 'BLQ':4, 'Rec':3, 'LwQ':2, 'Unf':1, 'NA':0}]
scale_mapper_BsmtFinType2 = ['BsmtFinType2', {'GLQ':6, 'ALQ':5, 'BLQ':4, 'Rec':3, 'LwQ':2, 'Unf':1, 'NA':0}]
scale_mapper_HeatingQC = ['HeatingQC', {'Ex':4, 'Gd':3, 'TA':2, 'Fa':1, 'Po':0}]
scale_mapper_KitchenQual = ['KitchenQual', {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'NA':0}]
scale_mapper_Functional = ['Functional', {'Typ':7, 'Min1':6, 'Min2':5, 'Mod':4, 'Maj1':3, 'Maj2':2, 'Sev':1, 'Sal':0}]
scale_mapper_FireplaceQu = ['FireplaceQu', {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'NA':0}]
scale_mapper_GarageQual = ['GarageQual', {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'NA':0}]
scale_mapper_GarageCond = ['GarageCond', {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'NA':0}]
scale_mapper_PoolQC = ['PoolQC', {'Ex':4, 'Gd':3, 'TA':2, 'Fa':1, 'NA':0}]
scale_mapper_GarageCond = ['GarageCond', {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'NA':0}]
scale_mapper_Fence = ['Fence', {'GdPrv':4, 'MnPrv':3, 'GdWo':2, 'MnWw':1, 'NA':0}]

# consolidate scale mappers to list and for loop onto column
scale_mapper_lst = [
    scale_mapper_ExterQual,
    scale_mapper_ExterCond,
    scale_mapper_BsmtQual,
    scale_mapper_ExterCond,
    scale_mapper_BsmtQual,
    scale_mapper_BsmtCond,
    scale_mapper_BsmtExposure,
    scale_mapper_BsmtFinType1,
    scale_mapper_BsmtFinType2,
    scale_mapper_HeatingQC,
    scale_mapper_KitchenQual,
    scale_mapper_Functional,
    scale_mapper_FireplaceQu,
    scale_mapper_GarageQual,
    scale_mapper_GarageCond,
    scale_mapper_PoolQC,
    scale_mapper_GarageCond,
    scale_mapper_Fence
]

for header, scale_mapper in scale_mapper_lst:
#     print(header)
#     print(scale_mapper)
    housing_data[header] = housing_data[header].replace(scale_mapper)
```

```python
# 4) dummy encode categorical nominal and drop original feature column
feature_dummy_lst = ['MSSubClass', 
                     'MSZoning', 
                     'Street',
                     'Alley',
                     'LotShape',
                     'LandContour',
                     'Utilities',
                     'LotConfig',
                     'LandSlope',
                     'Neighborhood',
                     'Condition1',
                     'Condition2',
                     'BldgType',
                     'HouseStyle',
                     'RoofStyle',
                     'RoofMatl',
                     'Exterior1st',
                     'Exterior2nd',
                     'MasVnrType',
                     'Foundation',
                     'Heating',
                     'CentralAir',
                     'Electrical',
                     'GarageType',
                     'GarageFinish',
                     'PavedDrive',
                     'MiscFeature',
                     'SaleType',
                     'SaleCondition'
                    ]

for feature in feature_dummy_lst:
    dummies = pd.get_dummies(housing_data[feature],
                             prefix=feature,
                             drop_first=False)
    housing_data = pd.concat([housing_data, dummies], axis=1).drop([feature], axis=1)

# rename to housing_sales and set index to 'PID'
housing_sales = housing_data.set_index('PID')
```

```python
# 5) import and build tax features
# load general real estate data info 
real_estate_data = pd.read_csv('Ames Real Estate Data.csv', low_memory=False)

# select Iowa ('IA') state data
real_estate_data_IA = real_estate_data[real_estate_data['MA_State']=='IA']

# filter for up to date info on duplicate rows
duplicate_lst = [
    'GeoRefNo',
    'Prop_Addr',
    'ClassPr_S',
    'ClassSc_S',
    'MA_Line1',
    'MA_Line2',
    'MA_City',
    'MA_State',
    'MA_Zip1',
    'Rcrd_Yr',
    'Rcrd_Mo',
    'LndAc_S',
    'OthAc_S',
    'TtlVal_AsrYr',
    'ValType',
    'Neighborhood',
    'LotArea',
    'BldgType',
    'YrBuilt',
    'HouseStyle',
    'Foundation',
    'RoofMatl',
    'Ext1',
    'Ext2',
    'MasVnrType',
    'Heating',
    'Central Air',
    'GLA',
    'TtlBsmtSF',
    'TotRmsAbvGrd',
    'Fireplaces',
    'PoolArea',
    'GarageType',
    'GarYrBlt',
    'Cars',
    'GarageArea']

tmp = real_estate_data_IA.sort_values(by = ['YrSold_YYYY', 'MoSold_MM'], 
                                    ascending = [False, False], 
                                    na_position = 'first').drop_duplicates(duplicate_lst)

# find max value by team, sort descending
# group by team and calculate mean, sum, and standard deviation of points
tmp2 = tmp.groupby(['GeoRefNo']).agg(GLA=('GLA', np.max), LotArea=('LotArea', np.max))
tmp2 = tmp2.reset_index()

# merge into real_estate_data_IA
real_estate_data_IA = pd.merge(tmp, tmp2, how='inner')

# rename columns in real_estate_data_IA
real_estate_data_IA = real_estate_data_IA.rename(columns={'MA_Zip1': 'zip_code', 'SalePrice': 'SalePrice2'})

# final feature selection in real_estate_data_IA
real_estate_data_features = real_estate_data_IA[['GeoRefNo', 'zip_code', 'TtlVal_AsrYr', 'GLA', 'SalePrice2', 'NmbrBRs']]

# final merge to sales date (housing_data)
real_estate_data_features = pd.merge(housing_data, real_estate_data_features,
                                   how='left',
                                   left_on=['PID'], right_on = ['GeoRefNo'])
```

```python
# load tax data
# source: https://www.irs.gov/statistics/soi-tax-stats-individual-income-tax-statistics-2020-zip-code-data-soi
# data was manually cleaned due to complexity of the file downloaded from IRS
tax_data = pd.read_csv('IA_Taxes.csv')

# feature engineering (refer to data dictionary for details)
tax_data['agi_per_ret']=tax_data['adj_gross_inc']/tax_data['total_returns']*1000
tax_data['perc_business_ret']=tax_data['business_income']/tax_data['adj_gross_inc']
tax_data['perc_farm_ret']=tax_data['farming_return']/tax_data['total_returns']
tax_data['perc_umemp_ret']=tax_data['unemp_returns']/tax_data['total_returns']
tax_data['perc_ssn_benefits']=tax_data['ssn_benefits']/tax_data['adj_gross_inc']
tax_data['perc_student_loans']=tax_data['student_loans']/tax_data['total_returns']
tax_data['perc_child_credits']=tax_data['child_credit_amt']/tax_data['adj_gross_inc']
tax_data['Perc_earned_inc_tax']=tax_data['earned_inc_tax']/tax_data['adj_gross_inc']

# select tax features for analysis
tax_data_features = tax_data[['zip_code', 
                              'total_returns', 
                              'adj_gross_inc',
                              'agi_per_ret', 
                              'perc_business_ret', 
                              'perc_farm_ret', 
                              'perc_umemp_ret', 
                              'perc_ssn_benefits', 
                              'perc_student_loans', 
                              'perc_child_credits', 
                              'Perc_earned_inc_tax']]

# merge real estate data with tax data features (note: merge is on zip code)
real_estate_tax_data_features = pd.merge(real_estate_data_features, tax_data_features, how='left')

# select features to include and set index
real_estate_tax_data_features = real_estate_tax_data_features[['PID', 
                                                               'TtlVal_AsrYr',
                                                               'GLA', 
                                                               'total_returns', 
                                                               'adj_gross_inc',
                                                               'agi_per_ret',
                                                               'perc_business_ret', 
                                                               'perc_farm_ret', 
                                                               'perc_umemp_ret',
                                                               'perc_ssn_benefits',
                                                               'perc_student_loans',
                                                               'perc_child_credits',
                                                               'Perc_earned_inc_tax']].set_index('PID')

# rename to housing_tax and replace missing values with mean feature value
housing_tax = real_estate_tax_data_features.fillna(real_estate_tax_data_features.mean())
```
